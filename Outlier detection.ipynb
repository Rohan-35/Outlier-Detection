# Finding the outliers

## Using IQR (inner quartile range)

```python
import pandas as pd
out_dataset = pd.read_csv('/content/Position_Salaries.csv') 
```

```python
out_dataset.head()
```

Let's find wheather any of the records in **Salary** column has any outlier using IQR

```python
out_dataset['Salary'].describe()  # to get a statistical overview of our target column
```

```python
percentile25 = out_dataset['Salary'].describe()[4]  #getting first quartile
percentile75 = out_dataset['Salary'].describe()[6]  #getting third quartile
iqr = out_dataset['Salary'].describe()[5]
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
# for counting number of total outliers:
count = int(out_dataset['Salary'].describe()[0])
n_outputs = 0
for i in range(count):
  if out_dataset['Salary'][i] > upper_limit:
    n_outputs = n_outputs+1
  elif out_dataset['Salary'][i] < lower_limit:
    n_outputs = n_outputs+1
print("No. of Outliers in the target column: ",n_outputs)
```

```python
# for getting thier positions..
count = int(out_dataset['Salary'].describe()[0])
for i in range(count):
  if out_dataset['Salary'][i] > upper_limit:
    print("position number(s) which have outliers: ", i)
  elif out_dataset['Salary'][i] < lower_limit:
    print("position number(s) which have outliers: ", i)
```

### Do note that we can get different results from box plot and IQR, we have to therefore use both these methods for coming into a valid conclusion.
